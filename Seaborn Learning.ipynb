```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips =  sns.load_dataset('tips')
tips
```

```python
sns.distplot(tips['total_bill'] ,  kde = False , bins = 40) #single variable plot
#gives histogram and kde
```

```python
#multi variable plot
sns.jointplot(x= , y= , data= )#2 distplots for bivarient data
```

```python
sns.jointplot(x='total_bill' , y='tip' , data= tips , kind = 'hex') #pass column names, kind  = style 
```

```python
sns.jointplot(x='total_bill' , y='tip' , data= tips , kind = 'reg')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data= tips , kind = 'kde')
```

```python
sns.pairplot(tips)#pairwise relationship accross the dataset for numerical columns
              # takes hue for categorical columns
```

```python
sns.pairplot(tips , hue = 'sex' )
```

```python
sns.rugplot(tips['total_bill']) #single column
```

```python
sns.kdeplot(tips['total_bill'])
```

### Categorical plots

```python
sns.barplot(x='sex' , y='total_bill' , data=tips)
```

```python
import numpy as np
sns.barplot(x='sex' , y='tip' , data=tips , estimator = np.std)
```

```python
#estimator counts number of occurences
sns.countplot(x='smoker' , data = tips)
```

```python
#box and whisker shows distribution
sns.boxplot(x= 'day',y ='total_bill' , data=tips)
```

```python
sns.boxplot(x= 'day',y ='total_bill' , data=tips , hue='smoker')
```

```python
#violin plot
#shows kde of underlying distribution
sns.violinplot(x='day' , y = 'total_bill' , data = tips)
```

```python
sns.violinplot(x='day' , y = 'total_bill' , data = tips , hue = 'smoker', split = True)
```

```python
sns.stripplot(x='day' , y='total_bill' , data = tips , hue='sex' , dodge = True)
```

```python
sns.violinplot(x='day' , y='total_bill' , data = tips , color='red' )
sns.swarmplot(x='day' , y='total_bill' , data = tips  , hue = 'sex')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data= tips)
```

```python
df = sns.load_dataset('diamonds')
df.head()
```

```python
pv1 = df.pivot_table(index = 'carat' , columns = 'clarity' , values = 'price')
fig, ax = plt.subplots(figsize=(10 , 30))
sns.heatmap(ax = ax , data = pv1)
```

```python
flights = sns.load_dataset('flights')
pv1 = flights.pivot_table(index = 'year' , columns = 'month' , values = 'passengers')
pv1
```

```python
sns.clustermap(pv1 , cmap = 'coolwarm' , linecolor='white' , linewidths = 2)
```

```python
sns.heatmap(df.corr() , linewidths = 3)
```
