```python
df = sns.load_dataset('diamonds')
df.head()
```

```python
pv1 = df.pivot_table(index = 'carat' , columns = 'clarity' , values = 'price')
fig, ax = plt.subplots(figsize=(10 , 30))
sns.heatmap(ax = ax , data = pv1)
```

```python
flights = sns.load_dataset('flights')
pv1 = flights.pivot_table(index = 'year' , columns = 'month' , values = 'passengers')
pv1
```

```python
sns.clustermap(pv1 , cmap = 'coolwarm' , linecolor='white' , linewidths = 2)
```

```python
sns.heatmap(df.corr() , linewidths = 3)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris) # creates empty grids
g.map(plt.scatter) #plot at diagonal
```

###  CLEAN GRAPH

```python
sns.set_context('notebook' , font_scale = 1.5)
ax = sns.countplot( x='sex' ,data =tips)
ax.set_yticklabels([])
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height-15,
            '{}'.format(height),
            ha='center' , va = 'baseline' , color = 'white') 
sns.despine(left = True , bottom = True)
```
