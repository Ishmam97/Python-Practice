```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0 , 5 , 11)
y = x ** 2

#functional
plt.plot(x, y , 'y+') 
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()# print the plot
```

```python
plt.subplot(1,2,1)
plt.plot(x , y , 'r-')

plt.subplot(1,2 , 2)
plt.plot(x , y , 'bo')
plt.show()
```

```python
# object orientated
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1 , 0.8 , 0.8])
axes1.plot(x, y)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Tit')
```

```python
axes2 = fig.add_axes([0.2 , 0.6 , 0.3 , 0.3])
axes2.plot(y,x)
```

```python
fig
```

```python
fig , axes = plt.subplots(nrows = 3 , ncols = 3)#specify rows n columns
# fix overlap
plt.tight_layout()
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols = 2)
for cur_ax in axes:
    cur_ax.plot(x , y)
```

```python
axes[1].cla() #cla() clears axes
axes[1].plot(y,x)
fig
```

```python
axes[0].set_title('First')
axes[1].set_title('Second')
fig
```

```python
fig.savefig('graphs.png')
```
