```python
import pdfplumber
import re
import numpy as np
import pandas as pd
```

```python
s = ''
with pdfplumber.open("input.pdf") as pdf:
    for i in range (len(pdf.pages)):
        s+= pdf.pages[i].extract_text()

print(s)
```

```python
lines = s.splitlines()
courses = []
for x in lines:
    aux = []
    if re.search("[A-Z]{4}", x):
        continue
    elif re.search("[A-Z]{3}\s\d{3}", x):
        aux.append(x[0:3])
        aux.append(x[4:7])
        aux.append(x[8:-2])
        aux.append(x[-1:])
        courses.append(aux)

df = pd.DataFrame(courses , columns =['Dept', 'Course Code', 'Course Name' , 'Credits']) 
df.reset_index(drop = True).set_index(['Dept' , 'Course Code'])
```
